```python
import pandas as pd
import numpy as np
```

# 1.修改行、列名称

### 直接修改

```python
df = pd.read_csv('sample.csv', index_col=0)
df
```

```python
df.columns
```

```python
df.columns = np.arange(6) # 可直接修改列名，行也是一样
df
```

### 使用rename函数修改

```python
df.rename(index={'a': 'xx', 'b': 'yy'}, inplace=True)
df
```

# 2.修改行、列的顺序

```python
df.reindex(['c', 'd', 'e', 'f', 'g', 'h', 'xx', 'yy']) # 使用reindex修改行的顺序
```

```python
df[[5, 4, 3, 2, 1, 0]] # 直接修改列的顺序
```
