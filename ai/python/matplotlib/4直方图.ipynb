```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']  # matplotlibÂä†ÂÖ•‰∏≠ÊñáÊîØÊåÅ
```

# 1.Âü∫Êú¨Áõ¥ÊñπÂõæ

```python
x = np.random.randn(1000)
```

binsÔºöüìäÊù°Êï∞

alphaÔºöÈÄèÊòéÂ∫¶

colorÔºöÈ¢úËâ≤

densityÔºöÈù¢ÁßØÊòØÂê¶ÂΩí‰∏∫1

```python
plt.hist(x, bins=20, alpha=0.5, color='steelblue', density=False)
plt.axvline(np.mean(x), c='red') # axvlineÊèíÂÖ•Á´ñÁ∫ø 
plt.xlabel("ÂÄº")
plt.ylabel("Êï∞Èáè")
plt.title("Ê≠£ÊÄÅÂàÜÂ∏ÉÊäΩÊ†∑")
plt.show()
```

# 2.Â§ö‰∏™histÁªÑÂêà

```python
x1 = np.random.normal(0, 1, 1000)
x2 = np.random.normal(3, 2, 1000)
x3 = np.random.normal(-2, 1.5, 1000)

kwargs = dict(bins=40, alpha=0.3, density=True) # Áî±kwargsË°®Á§∫histÂèÇÊï∞Ôºå‰∏çÂøÖÈÖçÁΩÆcolorÔºå‰ºöËá™Âä®Âå∫ÂàÜÂºÄ

plt.hist(x1, **kwargs, label="x1")
plt.hist(x2, **kwargs, label="x2")
plt.hist(x3, **kwargs, label="x3")
plt.legend()
plt.show()
```
